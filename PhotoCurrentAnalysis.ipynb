```python
from sympy import *
init_printing()
import numpy
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def loadData (fileName):
    u"""Загружает данные из файла.
        
        Loads data from the flie."""
    
    dataDir = 'E:\\My Documents\\work_docs\\work\\QT\\Homodyne\\PhotoCurrentData_20151019\\'
    with open(dataDir + fileName) as dataF1:
        dataList = numpy.array([[float(numString.replace(',','.')) for numString in line.split()] for line in dataF1])
    return dataList
```

```python
dataList = loadData('2.5M-3M.dat')
dataList1 = loadData('2.5M-3M-1.dat')
dataList2 = loadData('2.5M-3M-2.dat')
```

```python
def plotChannels(dataList, windowWidth = None):    
    fig, ax = plt.subplots(3,1)
    fig.set_figheight(11)
    fig.set_figwidth(15)
    ax[0].scatter(range(len(dataList[:windowWidth])),dataList[:windowWidth,0])
    ax[1].scatter(range(len(dataList[:windowWidth])),dataList[:windowWidth,1])
    ax[2].scatter(range(len(dataList[:windowWidth])),dataList[:windowWidth,2])
```

```python
plotChannels(dataList2[500:])
```

```python
plotChannels(dataList1[500:])
```

```python
plotChannels(dataList[500:])
```
